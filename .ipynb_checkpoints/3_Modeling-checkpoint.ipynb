# Modeling

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split, cross_val_score
#from sklearn.preprocessing import StandardScaler, PolynomialFeatures
#import statsmodels.api as sm
```

```python
kaggle_orig = pd.read_csv('./datasets/test.csv')
```

```python
#Function to perform most of the repetitive tasks in a linear regression

def regression(is_log, X, y):
    lr = LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(X, y)#, random_state=200)
    lr.fit(X_train, y_train)
    
    cv_score = cross_val_score(lr, X_train, y_train, cv=5).mean()
    r2_train = lr.score(X_train, y_train)
    r2_test = lr.score(X_test, y_test)
    
    y_preds = lr.predict(X_test)
    predictions = lr.predict(kaggle)

    if is_log:
        mae = mean_absolute_error(np.exp(y_test), np.exp(y_preds))
        mse = mean_squared_error(np.exp(y_test), np.exp(y_preds))
        rmse = np.sqrt(mse)
        
    else:
        mae = mean_absolute_error(y_test, y_preds)
        mse = mean_squared_error(y_test, y_preds)
        rmse = np.sqrt(mse)
    
    plt.figure(figsize=(10,6))
    if is_log:
        plt.scatter(np.exp(y_preds), np.exp(y_test), s=5, color='black', alpha = 0.8)
        plt.plot([0, np.max(np.exp(y_preds))], [0, np.max(np.exp(y_preds))], color = 'purple', linewidth=2)
    
    else:
        plt.scatter(y_preds, y_test, s=5, color='black', alpha = 0.8)
        plt.plot([0, np.max(y_preds)], [0, np.max(y_preds)], color = 'purple', linewidth=2)
        
    plt.xlabel("Predicted Values", fontsize = 18)
    plt.ylabel("Actual Values", fontsize = 18)
    plt.title('Predicted Values vs. Actual Values', fontsize = 22);
    
    if is_log:
        submission = pd.DataFrame(np.exp(predictions), columns=['SalePrice'])
    else:
        submission = pd.DataFrame(predictions, columns=['SalePrice'])
        
    submission['Id'] = kaggle_orig['Id']
    submission = submission.loc[:, ['Id', 'SalePrice']]
    
    print(f'Cross Validation Score: {cv_score}\nR2_train: {r2_train}\nR2_test: {r2_test}\nMean Absolute Error: {mae}\nMean Squared Error: {mse}\nRoot Mean Squared Error: {rmse}')
        
```

## 1. Model with the highest correlated features

```python
X = pd.read_csv('./datasets/X_corr.csv')
y = np.exp(pd.read_csv('./datasets/y.csv'))
kaggle = pd.read_csv('./datasets/kaggle_corr.csv')
```

```python
regression(0, X, y)
```

```python
submission.to_csv('./submissions/corr.csv', index=False)
```
